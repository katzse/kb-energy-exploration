```python
import os
import pandas as pd
import numpy as np
import math
```

```python
# Change directory 
os.chdir("C:/Kate/Bioinformatics/Diplom")

# Assign spreadsheet filename to `file`
file = 'Selection_434.xlsx'

# Load spreadsheet
xf = pd.ExcelFile(file)

# Print the sheet names
print(xf.sheet_names)

# Load a sheet into a DataFrames by name: dfSelection
dfSelection = xf.parse('Selection')

# Convert names of proteins in dataFrame to list
names = dfSelection["ID"].tolist()
```

```python
# Change directory 
os.chdir("C:/Kate/Bioinformatics/Diplom/Set")

# Calculate parameters for the whole dataset (200 proteins):

path = 'C:/Kate/Bioinformatics/Diplom/Set/'
f = '.xlsx'
seqLen = 0
dbLen = 0
dfWhole = pd.DataFrame(columns=['Index','Residue','sumSASA','rSASA','maxProtOr','RSA','b0e1'])
#print(dfWhole)

for i in names:
    protFile = i+f
    
    # Load spreadsheet
    xf = pd.ExcelFile(protFile)
    
    # Load a sheet into a DataFrames by name of the protein
    dfData = xf.parse('data')
    
    dfWhole = dfWhole.append(dfData, ignore_index=True)
```

```python
# Calculate the Length of the data set
Length = len(dfWhole)
Length
```

```python
# Count residue types 
# R = RE + RB
numR = dfWhole.groupby('Residue')['b0e1'].count()
```

```python
numRB = dfWhole[dfWhole['b0e1'] == 0].groupby('Residue')['b0e1'].count()
```

```python
numRE = dfWhole[dfWhole['b0e1'] == 1].groupby('Residue')['b0e1'].count()
```

```python
FRB = numRB/Length
```

```python
FRE = numRE/Length
```

```python
# Count buried and exposed residues
numBE = dfWhole.groupby('b0e1')['Residue'].count()
numBE
```

```python
expPB = numBE.at[0]/Length
expPE = numBE.at[1]/Length
print('expPB', expPB)
print('expPE', expPE)
expPB+expPE
```

```python
PobsRB = numRB/numR
```

```python
PobsRE = numRE/numR
```

```python
# LOG base 10 - 'log10'
KBenergyRB_log10 = []
for p in PobsRB:
    KBenergyRB_log10.append(math.log10(expPB/p))
```

```python
# LOG base 10 - 'log10'
KBenergyRE_log10 = []
for p in PobsRE:
    KBenergyRE_log10.append(math.log10(expPE/p))
```

```python
# LOG with base 'e' (ln) !!  
KBenergyRB = []
for p in PobsRB:
    KBenergyRB.append(math.log(expPB/p))
```

```python
# LOG with base 'e' (ln) !!  
KBenergyRE = []
for p in PobsRE:
    KBenergyRE.append(math.log(expPE/p))
```

```python
dfParams = pd.DataFrame()
dfParams['Residue'] = numR.index
dfParams['numR'] = numR.values
dfParams['RB'] = numRB.values
dfParams['RE'] = numRE.values
dfParams['F_RB'] = FRB.values
dfParams['F_RE'] = FRE.values
dfParams['PobsRB'] = PobsRB.values
dfParams['PobsRE'] = PobsRE.values
dfParams['KBenergyRB'] = KBenergyRB
dfParams['KBenergyRE'] = KBenergyRE
dfParams['KBenergyRB_log10'] = KBenergyRB_log10
dfParams['KBenergyRE_log10'] = KBenergyRE_log10

dfParams
```

```python
dfKBES = pd.DataFrame()
dfKBES['Residue'] = numR.index
dfKBES['F_RB'] = FRB.values
dfKBES['F_RE'] = FRE.values
dfKBES['KBenergyRB'] = KBenergyRB
dfKBES['KBenergyRE'] = KBenergyRE
dfKBES
```

```python
# Change directory 
os.chdir("C:/Kate/Bioinformatics/Diplom/Resample/")
writer = pd.ExcelWriter(('FullSet.xlsx'), engine='xlsxwriter')

# Write your DataFrame to a file     
dfWhole.to_excel(writer, 'FullSet')
dfKBES.to_excel(writer, 'KBES')
dfParams.to_excel(writer, 'Params')
numBE.to_excel(writer, 'numBE')

# Save the result 
writer.save()    
```
