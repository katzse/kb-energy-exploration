```python
import os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Open and load data
# Change directory 
os.chdir("C:/Kate/Bioinformatics/Diplom/Results")

# Assign spreadsheet filename to `file`
file = 'Params_FullSet.xlsx'
# Load spreadsheet
xf = pd.ExcelFile(file)
# Load a sheet into a DataFrames by name: dfSelection
dfValues = xf.parse('KBES')
```

```python
def createPlot(shift, dfValues): 
    
    os.chdir("C:/Kate/Bioinformatics/Diplom/Resample20x100_Shift/")
    # Assign spreadsheet filename to `file`
    file = 'StatParams20x100_shift'+str(shift)+'.xlsx'
    # Load spreadsheet
    xf = pd.ExcelFile(file)
    # Load a sheet into a DataFrames by name: dfSelection
    dfMean2 = xf.parse('mean')
    dfStd2 = xf.parse('std')
    dfVar2 = xf.parse('var')
    
    os.chdir("C:/Kate/Bioinformatics/Diplom/Resample50x100_Shift/")
    # Assign spreadsheet filename to `file`
    file = 'StatParams50x100_shift'+str(shift)+'.xlsx'
    # Load spreadsheet
    xf = pd.ExcelFile(file)
    # Load a sheet into a DataFrames by name: dfSelection
    dfMean5 = xf.parse('mean')
    dfStd5 = xf.parse('std')
    dfVar5 = xf.parse('var')
    
    # Frequency of Buried Residues - mean, std for sample sizes 20 and 50

    N = len(dfMean2.Residue)
    ind = np.arange(N) # the x locations for the groups
    width = 0.35       # the width of the bars
    frb, ax = plt.subplots()
    graphFRB2 = ax.bar(ind, dfMean2.F_RB, width, color='m', yerr = dfStd2.F_RB, capsize=2)
    graphFRB5 = ax.bar(ind + width, dfMean5.F_RB, width, color='c', yerr = dfStd5.F_RB, capsize=2)

    # Add text for labels, title and axes ticks
    ax.set_ylabel('mean, r.u.')
    ax.set_title('Frequency of Buried Residues, shift '+str(shift))
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels((dfMean2.Residue))

    ax.legend((graphFRB2[0], graphFRB5[0]), ('Sample size 20', 'Sample size 50'))

    # Plot the vlues from the whole data set 
    plt.plot(ind, dfValues.F_RB, '_', color='grey')
    
    

    # Frequency of Expozed Residues - mean, std for sample sizes 20 and 50
    
    N = len(dfMean2.Residue)
    ind = np.arange(N) # the x locations for the groups
    width = 0.35       # the width of the bars
    fre, ax = plt.subplots()
    graphFRE2 = ax.bar(ind, dfMean2.F_RE, width, color='m', yerr = dfStd2.F_RE, capsize=2)
    graphFRE5 = ax.bar(ind + width, dfMean5.F_RE, width, color='c', yerr = dfStd5.F_RE, capsize=2)

    # Add text for labels, title and axes ticks
    ax.set_ylabel('mean, r.u.')
    ax.set_title('Frequency of Exposed Residues, shift '+str(shift))
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels((dfMean2.Residue))

    ax.legend((graphFRE2[0], graphFRE5[0]), ('Sample size 20', 'Sample size 50'))

    # Plot the vlues from the whole data set 
    plt.plot(ind, dfValues.F_RE, '_', color='grey')
    
    
    # KB Energy of Buried Residues - mean, std for sample sizes 20 and 50

    N = len(dfMean2.Residue)
    ind = np.arange(N) # the x locations for the groups
    width = 0.35       # the width of the bars
    kbeb, ax = plt.subplots()
    graphKBEB2 = ax.bar(ind, dfMean2.KBenergy_RB, width, color='g', yerr = dfStd2.KBenergy_RB, capsize=2)
    graphKBEB5 = ax.bar(ind + width, dfMean5.KBenergy_RB, width, color='y', yerr = dfStd5.KBenergy_RB, capsize=2)

    # Add text for labels, title and axes ticks
    ax.set_ylabel('mean, r.u.')
    ax.set_title('KB Energy of Buried Residues, shift '+str(shift))
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels((dfMean2.Residue))

    ax.legend((graphKBEB2[0], graphKBEB5[0]), ('Sample size 20', 'Sample size 50'))

    # Plot the vlues from the whole data set 
    plt.plot(ind, dfValues.KBenergyRB, '_', color='grey')
    
    
    # KB Energy of Exposed Residues - mean, std for sample sizes 20 and 50

    N = len(dfMean2.Residue)
    ind = np.arange(N) # the x locations for the groups
    width = 0.35       # the width of the bars
    kbee, ax = plt.subplots()
    graphKBEE2 = ax.bar(ind, dfMean2.KBenergy_RE, width, color='g', yerr = dfStd2.KBenergy_RE, capsize=2)
    graphKBEE5 = ax.bar(ind + width, dfMean5.KBenergy_RE, width, color='y', yerr = dfStd5.KBenergy_RE, capsize=2)

    # Add text for labels, title and axes ticks
    ax.set_ylabel('mean, r.u.')
    ax.set_title('KB Energy of Exposed Residues, shift '+str(shift))
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels((dfMean2.Residue))

    ax.legend((graphKBEE2[0], graphKBEE5[0]), ('Sample size 20', 'Sample size 50'))

    # Plot the vlues from the whole data set 
    plt.plot(ind, dfValues.KBenergyRE, '_', color='grey')
    
    
    os.chdir("C:/Kate/Bioinformatics/Diplom/Images/Shift/") 
    frb.savefig('frb_shift'+str(shift)+'.png', dpi=1000)
    fre.savefig('fre_shift'+str(shift)+'.png', dpi=1000)
    kbeb.savefig('kbeb_shift'+str(shift)+'.png', dpi=1000)
    kbee.savefig('kbee_shift'+str(shift)+'.png', dpi=1000)

    
    return plt
```

```python
for s in range(-5, 6):
    plt = createPlot(s, dfValues)
```
